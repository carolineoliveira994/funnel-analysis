Embora o Google Ads apresente o maior ROI médio, a mediana significativamente menor indica que a maioria das campanhas possui desempenho inferior à média.

Isso sugere que poucos outliers estão puxando o resultado para cima.

Recomendação:
Identificar campanhas de alto desempenho e replicar estratégias para aumentar consistência.


```python
Q1 = df['ROI'].quantile(0.25)
Q3 = df['ROI'].quantile(0.75)

IQR = Q3 - Q1

limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

outliers = df[(df['ROI'] < limite_inferior) | (df['ROI'] > limite_superior)]

outliers.sort_values(by='ROI', ascending=False).head(10)
```

As campanhas com maior ROI apresentam um padrão claro de baixo custo aliado a alto volume de conversões, resultando em receitas significativamente superiores ao investimento.

Observa-se que o principal driver de ROI elevado é a alta taxa de conversão, mais do que volume de cliques ou impressões.

Adicionalmente, foram identificados possíveis outliers inconsistentes (ex: conversões superiores ao número de cliques), indicando limitações na qualidade dos dados.

Focar em estratégias que maximizem conversão com baixo custo e implementar validação de dados para garantir consistência analítica.
